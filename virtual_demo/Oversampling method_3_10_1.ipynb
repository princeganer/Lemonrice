```python
import pandas as pd
```

```python
bank_df = pd.read_csv("oversampling_bank_full_3_10_1.csv",
                      index_col = 0)
bank_df.head()
```

```python
display(bank_df["y"].value_counts())
```

```python
def over_sample( data, oversampling_ratio ):
    import math

    # avoid changing the original object accidentally
    new_data = bank_df.copy()

    nums = new_data['y'].value_counts()
    num_n = nums[0]
    num_y = nums[1]
    nums = len(new_data["y"])

    new_y = ( (1 - oversampling_ratio) / oversampling_ratio)*num_n

    big_loop = list(str( (new_y / num_y) ).split( sep = "."))
    loop_num = int( math.ceil(new_y/num_y) )

    new_df = new_data[ new_data["y"] == "yes" ]

    for i in range(0, loop_num-1):
        # randomly select all rows from new_df
        random_rows = new_df.sample(n = num_y, replace=True)

        # append the selected row to bank_df
        new_data = new_data.append(random_rows, ignore_index=True)
    
    return new_data
```

```python
bank_oversampled_df = over_sample(bank_df, 0.6)
```
