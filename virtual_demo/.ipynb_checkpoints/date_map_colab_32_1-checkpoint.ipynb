```python
def map_year(data, start_yr, end_yr):
    month_lst = ["jan", "feb", "mar", "apr", "may", "jun", "jul", "aug", "sep", "oct", "nov", "dec"]

    # Make a copy of the original dataframe
    new_data = data.copy()

    # Insert a new "year" column filled with zeros
    new_data.insert(loc=10, column="year", value=0)

    # Set the first year to the start year
    current_year = int(start_yr)
    new_data.at[0, "year"] = current_year

    # Loop through the rows of the dataframe, updating the year column when the month changes
    for i in range(1, len(new_data)):
        # If the current month is earlier in the year than the previous month, increment the year
        if month_lst.index(new_data["month"][i]) < month_lst.index(new_data["month"][i-1]):
            current_year += 1

        new_data.at[i, "year"] = current_year

        # If the current year exceeds the end year, break out of the loop
        if current_year > end_yr:
            break

    return new_data
```

```python
result_df = map_year( data = df_add, start_yr = 2008, end_yr = 2010)
```

```python
import calendar
```

```python
def get_day_num(year, month, day):
    month_lst = [ "haha",
               "jan", "feb", "mar", "apr",
               "may", "jun", "jul", "aug",
               "sep", "oct", "nov", "dec"]
    month_num = int(month_lst.index(month))
    return calendar.monthrange(year, month_num)[1] - calendar.monthrange(year, month_num)[0]
```

```python
# This is creating a problem, every day is mapped same date
print(get_day_num(2008, "mar", "sun"))
print(get_day_num(2008, "mar", "mon"))
print(get_day_num(2008, "mar", "tue"))

# This is creating a problem, every day is mapped same date
print(get_day_num(2008, "apr", "sun"))
print(get_day_num(2008, "may", "sun"))
print(get_day_num(2008, "mar", "tue"))
```

Proposed Solution: </b>
Give the first day_of_week fetched from dataset for the month the first week's date of the repective day and consecutive days will get the remaining as the data is ordered like the months.

```python
def map_date(data):
    # Make a copy of the original dataframe
    new_data = data.copy()
    
    # Insert a new "date" column filled with zeros
    new_data.insert(loc=10, column="date", value=0)
    
    month_lst = [ "haha",
               "jan", "feb", "mar", "apr",
               "may", "jun", "jul", "aug",
               "sep", "oct", "nov", "dec"]
    
    # Loop through the rows of the dataframe, updating the date column
    for i in range(0,len(new_data)):
        
        new_data.at[i, "date"] = str(new_data["year"][i]) + "-" + str( month_lst.index(new_data["month"][i]) )+ "-" + str( get_day_num(year = new_data["year"][i],month = new_data["month"][i], day = new_data["day_of_week"][i]) )
    return new_data
```
