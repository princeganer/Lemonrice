```python
def year_mapper( data, start_yr ):

  month_lst = [ "haha",
               "jan", "feb", "mar", "apr",
               "may", "jun", "jul", "aug",
               "sep", "oct", "nov", "dec"]

  # for storing first row's month year
  yr = int(start_yr)

  # for making a new dataframe same as passed dataframe
  new_data = data
    
  # for inserting a "year" column with value 0 at 10th location in new_data df
  new_data.insert(loc = 10, column = "year", value = int(0))
    
  # for assigning "year" column int datatype
  new_data.astype( {"year"  : "int"} )

  # for storing first month's index of dataframe
  current_month_idx = month_lst.index( data["month"][0] )
    
  # For iterating through rows, skipping the 1st row
  start_row_no = 1
  end_row_no = int(df_add.count()[0]) # end_row_no will be 1 more than actual end row number

  # Assigning first row's year as start_yr 
  new_data.at[0, "year"] = yr
    
  for row_no in range(start_row_no, end_row_no ):
        
    # giving flags for current year's current month exhausted or not
    if row_no != (end_row_no - 1):
        current_month = month_lst[current_month_idx]
        running_month = new_data["month"][row_no+1]
      
      if current_month == running_month:
        month_exhaust_flag = False 
      else:
        month_exhaust_flag = True
        check_yr = yr
    else:
      pass

    # loc on month and compare with next row's month
    # if both same then month remain same, month_exhaust_flag = False
    # if both diffferent then month change, month_exhaust_flag = True
    # we use this flag to avoid next year conflict, two consecutive year beginning from same month
    # month_exhaust_flag signifies wether current year's month is exhuasted or not  
    
    # for storing running month's index of dataframe
    running_month_idx = month_lst.index( data["month"][row_no] )
    
    if not month_exhaust_flag:
      if running_month_idx >= current_month_idx:
        new_data.at[row_no, "year"] = yr
    
      elif running_month_idx < current_month_idx:
        yr = yr + 1
        current_month_idx = month_lst.index( new_data["month"][row_no] )
        new_data.at[row_no, "year"] = yr
    else:
      if running_month_idx > current_month_idx:
        new_data.at[row_no, "year"] = yr
    
      elif running_month_idx < current_month_idx:
        yr = yr + 1
        current_month_idx = month_lst.index( new_data["month"][row_no] )
        new_data.at[row_no, "year"] = yr
    
  return new_data
                    
```
